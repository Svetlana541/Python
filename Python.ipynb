### Задание 11: 
Создайте функцию для выполнения основных арифметических операций, которая применяет сложение, вычитание, умножение и деление к строковому значению (например, "12 + 24" или "23-21" или "12 // 12" или "12 * 21").

Здесь у нас есть 1 число, за которым следует пробел, затем оператор, за которым следует другой пробел, и 2 число. Возвращаемое значение должно быть числом.

Применение функции eval() не допускается. В случае деления, всякий раз, когда второе число равно "0", возвращайте -1.

```python
def functionCalc(expression):
    arr = expression.split(" ")
    if arr[1] == "+":
        return int(arr[0]) + int(arr[2])
    if arr[1] == "-":
        return int(arr[0]) - int(arr[2])
    if arr[1] == "*":
        return int(arr[0]) * int(arr[2])
    if arr[1] == "//":
        if int(arr[2]) != 0:
            return int(arr[0]) // int(arr[2])
        else:
            return -1

print(functionCalc("15 // 0"))
print(functionCalc("15 // 5"))
print(functionCalc("15 + 5"))
print(functionCalc("5 * 5"))
print(functionCalc("15 - 5"))
```

* \# = 5
* О = 3
* Х = 1
* ! = -1
* !! = -3
* !!! = -5

Если итоговый результат отрицательный, верните 0 (например, 3 ``#``, 3 ``!!``, 2 ``!!!`` и ``X`` будет (5 + 5 + 5 - 3 - 3 - 3 - 5 - 5 + 1) = -3, так что верните 0.

```python
dict = {"#": 5, "O": 3, "X": 1, "!": -1, "!!": -3, "!!!": -5}
def functionList(value):
    sum = 0
    for i in value:
        for j in i:
            if j in dict: 
                sum += dict[j]
    if sum < 0:
        return 0
    else:
        return sum
    
print(functionList([["#", "#"],["X","X","X"],["!!"],["!!!", "!!!"]]))
print(functionList([["#", "#", "#", "#"],["X","X","X"],["!!"],["!!!", "!!!"]]))
```

Алгоритм такой же - первое число перекодируется в соответствием с таблицей ascii, второй символ - сумма первого и второго числа перекодированная с помощью таблицы ascii.

```python
def functionDecoder(data):
    if isinstance(data, str): #проверяем является ли переменная строкой
        res = []
        for i, v in enumerate(data):
            if i == 0: #код буквы в ascii
                res.append(ord(v)) #возвращаем числовое представление для указанного символа 
                continue
            res.append(ord(v) - ord(data[i-1])) #разница между текущим и предыдущим символом
            
    if isinstance(data, list): #проверяем является ли переменная списком
        res = ""
        for i, v in enumerate(data):
            if i == 0: #первое число перекодируется в соответствием с таблицей ascii
                res += chr(v) #возвращаем символ по числовому представлению
                continue
            res += chr(v + data[i-1]) #сумма первого и второго числа перекодированная с помощью таблицы ascii
            data[i] = v + data[i-1]
    return res

print(functionDecoder([72, 29, 7, 0, 3]))
print(functionDecoder("Hello"))
print(functionDecoder([ 72, 33, -73, 84, -12, -3, 13, -13, -68 ]))
```

Каждый может увидеть сцену в примере ниже:

``[[1, 2, 3, 2, 1, 1],
 [2, 4, 4, 3, 2, 2],
 [5, 5, 5, 5, 4, 4],
 [6, 6, 7, 6, 5, 5]]``

Не все видят сцену:

``[[1, 2, 3, 2, 1, 1], 
  [2, 4, 4, 3, 2, 2], 
  [5, 5, 5, 10, 4, 4], 
  [6, 6, 7, 6, 5, 5]]``

Функция должна возвращать True, если абсолютно все видят сцену, ичане False

```python
def functionSeat(dataList):
    for i, v in enumerate(dataList): #перебираем каждый элемент списка с учетом индекса
        if(i == 0): #пропускаем первый элемент основного списка и переходим к следующему, т.к. впереди него элементов нет (сравнивать не с чем) 
            continue
            
        for j, k in enumerate(dataList[i]): #перебираем каждый вложенный элемент списка
            if k < dataList[i-1][j]: #если текущий элемент меньше элемента из предыдущего списка, возвращаем false
                return False
    return True
            
print(functionSeat([[1, 2, 3, 2, 1, 1], [2, 4, 4, 3, 2, 2], [5, 5, 5, 5, 4, 4], [6, 6, 7, 6, 5, 5]]))

print(functionSeat([[1, 2, 3, 2, 1, 1], [2, 4, 4, 3, 2, 2], [5, 5, 5, 10, 4, 4], [6, 6, 7, 6, 5, 5]]))
```

########

____####

Замечания:
Возвращаемая строка дополняется символом перехода на новую строку \n

```python
def functionStaircase(value):
    result = ""
    if value > 0:
        for i in range(value):
            result += "_" * (value - (i+1)) #если число полож., повторяем '_' -> 5 - (i + 1) раз
            result += "#" * (i+1) + "\n" #
    else:
        for i in range(-value): #переводим в полож. число
            result += "_" * (i) #чтобы лестница была вниз, начинаем с '_'
            result += "#" * (-(value+i)) + "\n" 
    return result
print(functionStaircase(-5))
print('\n')
print(functionStaircase(5))
```

```python
def functionNearestChapter(dict, page):
    diff = page
    res = ""
    for i in dict: #переберем словарь по ключу
        if diff >= abs(dict[i] - page): #сравниваем разницу между страницей и значением словаря по ключу (abs возвращает значение по модулю)
            diff = abs(dict[i] - page)
            res = i

    return res

print(functionNearestChapter({
  "Chapter 1" : 1,
  "Chapter 2" : 15,
  "Chapter 3" : 37
}, 10))

print(functionNearestChapter({
  "New Beginnings" : 1,
  "Strange Developments" : 62,
  "The End?" : 194,
  "The True Ending" : 460
}, 200))

print(functionNearestChapter({
  "Chapter 1a" : 1,
  "Chapter 1b" : 5
}, 3))
```
